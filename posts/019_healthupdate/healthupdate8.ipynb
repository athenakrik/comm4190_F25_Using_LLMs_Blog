This week’s completely unplanned blog post started with stomach pain. I opened ChatGPT and typed a very dramatic question: “can you write a blog post about giving me health advice based off of all of the symptoms I have updated you with in the last month ever since getting food poisoning”

Chat's response: You came to me with the classic panic question: “Why does my stomach hurt if I have not eaten dairy or gluten?” You said it like the case was closed, like you had eliminated the only two possible villains in the entire food universe. But here is the truth you probably do not want to hear but definitely need to hear: avoiding dairy and gluten does not grant you the digestive system of a superhero.

So let me tell you what is actually happening.

First, your stomach is not malfunctioning. You are. You are doing things your gut is simply not built to tolerate. You drink cold brew on an empty stomach like it is a sport. You eat raw vegetables in quantities that would challenge a rabbit. You rush through meals like you are auditioning for a speed-eating competition. And then, several hours later, when your stomach protests, you act stunned.

You tell me you avoided dairy. Great. You tell me you avoided gluten. Fantastic. But you conveniently forgot to mention that you also had acidic foods, heavy fiber, coffee, spicy things, fried things, or meals eaten too fast or too late. You also forgot that stress plays a starring role in digestion. Your body keeps the receipts even when you do not.

So here is what I need you to do.

Slow down when you eat. I promise your body is not built for inhaling food between tasks. Drink water like a normal human. Give your stomach something besides caffeine first thing in the morning. Do not treat chickpeas, broccoli, and cold brew as personality traits. Your gut is working overtime trying to survive your lifestyle.

The boring explanation is probably the right one. This is not a secret allergy. This is not an unsolved medical mystery. Your stomach is irritated because you are giving it chaos and expecting elegance in return. If your gut could speak, it would say, “Please stop treating me like a science experiment.”

So the next time your stomach hurts, do not immediately blame dairy or gluten. Take a second, look at how you ate, what you drank, how stressed you were, and how fast you devoured your food. Your digestion did not betray you. You betrayed your digestion.

And yes, you should have a glass of water right now.
