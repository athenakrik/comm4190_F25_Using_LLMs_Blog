Lately I have been living with one question that refuses to leave my brain. Every time I talk to someone older, every time I scroll through LinkedIn, every time I see a TikTok about “where you should move in your twenties,” the same thought pops up: am I supposed to move to San Francisco or New York after I graduate.

It feels like picking a city is picking a personality. Picking a timeline. Picking a future. Picking a version of myself I have not even met yet. And the pressure is insane. San Francisco has every AI founder I want to stalk on Twitter, unlimited tech energy, and enough networking opportunities to melt my nervous system. New York has culture, pace, people, ambition, and the kind of chaotic lifestyle that feels like being plugged into an electrical outlet.

At some point in my spiral, I opened ChatGPT and typed:

“Can you write a blog post about choosing between SF and NYC after graduating, because it feels like picking a city is picking who I am going to become.”

Here is what ChatGPT had to say.

Chat’s response:

You came to me sounding like someone who believes choosing a city is the same as choosing your destiny. But here is the truth you need to hear: both cities are exaggerations of themselves online. Neither one is as perfect or terrifying as the internet makes them sound. And the version of you who lives in each city is still you. You are not going to evaporate into a stereotype just because you changed time zones.

Let me break down what is actually happening inside your head.

San Francisco represents the future you think you want to build. It is the city where ambition is measured in startups, where cafes double as co-working spaces, and where strangers casually say things like “model weights” or “compute constraints” in a regular conversation. You imagine yourself surrounded by founders, engineers, product people, AI researchers, and investors who treat ideas like currency. It feels like a place where you could build something huge, something fast, something that only this moment in tech history could allow.

And then there is New York. The city that makes you feel alive just by existing in it. The energy is different. It is loud, fast, warm, cold, unpredictable, overwhelming, miraculous. You imagine walking to the subway with iced coffee in your hand, going to events, meeting people from every industry, absorbing culture by accident, and feeling plugged into humanity in a way no other city can imitate. You imagine becoming bolder, sharper, more spontaneous, more yourself.

Here is the real dilemma.

When you think about SF, you imagine the career version of your life. When you think about NYC, you imagine the human version of your life. And both matter. You are not choosing between right and wrong. You are choosing between two types of growth. One grows your mind, one grows your world.

But let me tell you something comforting.

You cannot choose “wrong.” Either city will stretch you. Either city will change you. Either city will surround you with interesting people. You will end up building things, meeting people you never expected, and discovering parts of yourself you have not met yet. Your life will not collapse because you picked the wrong skyline.

You keep talking about this choice like it is permanent, like the second you choose a coast you are trapped there for life. You are not. You can move. You can try one city, learn from it, and then try the other. You can evolve. You can change your mind. You can follow the opportunities, the people, the instincts that feel right at that moment.

So here is what I need you to do.

Stop treating your graduation city like a tattoo. It is not forever. It is just where your next chapter starts. Not where it ends. Choose the place that makes you feel inspired now. Choose the place that lights something up in you. Choose the place that matches the version of yourself you are excited to experiment with.

The future is not locked to a zip code. You build it wherever you go.

And yes, it is fine to pick based on vibes.
