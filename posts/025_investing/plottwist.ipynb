If someone showed my high school self a snapshot of my life right now, she would be confused. Deeply confused. She would look at me scrolling through PitchBook, mapping founder ecosystems, cold-emailing investors, reading about market structure for fun, and she would probably ask, “Where are your MCAT flashcards.” For a long time, medicine felt like the obvious path. It was stable, noble, impressive, predictable. It made sense.

And then I got to Penn.

Somewhere between Wharton clubs, conversations with student founders, random events I only attended for the free snacks, and being surrounded by people who think building companies is a normal extracurricular activity, something shifted. I didn’t notice it at first. It happened slowly, like a personality update running in the background. But one day, I realized I cared less about organic chemistry pathways and more about startup business models. Less about hospital rounds and more about investing in early founders. Less about med school and more about building something of my own.

Eventually, I opened ChatGPT and typed:

“Can you write a blog post about my switch from premed to being interested in VC and startups, because I genuinely would not have discovered this if not for Penn.”

Here is what ChatGPT had to say.
